### EX2: Obstacle Avoidance

```python
import random
import math
from math import pi
import numpy as np
```

### Setup q1

```python
k = 1000
q = np.random.uniform(0.0, 15.0, k)
```

#### Q1

```python
from scipy.stats import *

def pretend_function(x):
    return (0.3 * norm.pdf(x, 2.0, 1.0) + 0.4 * norm.pdf(x, 5.0, 2.0) + 0.3 * norm.pdf(x, 9.0, 1.0)) / (1/15)

def axis_prob(x):
    return (0.3 * norm.pdf(x, 2.0, 1.0) + 0.4 * norm.pdf(x, 5.0, 2.0) + 0.3 * norm.pdf(x, 9.0, 1.0))

vectorized_pretend_func = np.vectorize(pretend_function)
vectorized_axis_prob = np.vectorize(axis_prob)
```

```python
# k = 20
weights_k20 = vectorized_pretend_func(q[0:20])
new_probabilities_k20 = weights_k20 / sum(weights_k20)
resample_k20 = np.random.choice(q[0:20], 20, True, new_probabilities_k20)

# k = 100
weights_k100 = vectorized_pretend_func(q[0:100])
new_probabilities_k100 = weights_k100 / sum(weights_k100)
resample_k100 = np.random.choice(q[0:100], 100, True, new_probabilities_k100)

# k = 1000
weights_k1000 = vectorized_pretend_func(q)
new_probabilities_k1000 = weights_k1000 / sum(weights_k1000)
resample_k1000 = np.random.choice(q, k, True, new_probabilities_k1000)

#new_probabilities = list(new_probabilities)
#print(new_probabilities)
#print(np.sort(resample))
#print("\n")
```

```python
import matplotlib.pyplot as plt

# Plots the expected prob distribution that we hope the histogram will follow
x_axis = np.linspace(0, 15, 100)
axis_prob = vectorized_axis_prob(x_axis)
plt.plot(x_axis, axis_prob, color="red", label="test")
plt.xticks(np.arange(0,15,1))

plt.hist(x=resample_k1000, density=True) # Plots the histogram of the resample
plt.title("Resample k=1000")
#plt.xlabel("range")
#plt.ylabel("frequency")
plt.show()
```

#### Q2

```python
q2 = np.random.normal(loc=5, scale=4, size=k)

weights_2_k20 = vectorized_pretend_func(q2[0:20])
new_probabilities_2_k20 = weights_2_k20 / sum(weights_2_k20)
resample_2_k20 = np.random.choice(q2[0:20], k, True, new_probabilities_2_k20)

weights_2_k100 = vectorized_pretend_func(q2[0:100])
new_probabilities_2_k100 = weights_2_k100 / sum(weights_2_k100)
resample_2_k100 = np.random.choice(q2[0:100], k, True, new_probabilities_2_k100)

weights_2_k1000 = vectorized_pretend_func(q2)
new_probabilities_2_k1000 = weights_2_k1000 / sum(weights_2_k1000)
resample_2_k1000 = np.random.choice(q2, k, True, new_probabilities_2_k1000)
```

```python
x_axis = np.linspace(0, 15, 100)
axis_prob = vectorized_axis_prob(x_axis)
plt.plot(x_axis, axis_prob, color="red", label="test")
plt.xticks(np.arange(0,15,1))

plt.hist(x=resample_2_k1000, density=True) # Plots the histogram of the resample

plt.title("Resample k = 1000 (norm dist)")
#plt.xlabel("range")
#plt.ylabel("frequency")
plt.show()
```
